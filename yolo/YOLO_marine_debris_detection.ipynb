### Импорты

```python
from ultralytics import YOLO
import torch
```

### Импорт диска

```python
from google.colab import drive
drive.mount('/content/drive')
```

### Обучение

```python
torch.cuda.empty_cache()
```

```python
model = YOLO('yolo12m.pt')

model.train(
    data='/content/drive/MyDrive/MarineDebrisDetection/data.yaml',
    epochs=50,
    imgsz=640,
    batch=8,
    name='yolo12m_custom',
    device=0,
    workers=2,
    cache=False
)
```

```python
torch.cuda.empty_cache()
```

### Оценка качества модели

```python
metrics = model.val()
```

### Проверка модели

```python
results = model("1.jpg")
results[0].show()
results[0].save("output.jpg")
```
